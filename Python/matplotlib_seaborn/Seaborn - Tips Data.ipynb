```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline 
```

```python
tips = pd.read_csv('tips.csv')
```

```python
tips.head()
```

### Regression Plots

```python
def labels_and_title(xlab=[], ylab=[], title=[], sns_multi=None):
    if sns_multi == None:
        plt.xlabel(xlab[0], fontsize=xlab[1])
        plt.ylabel(ylab[0], fontsize=ylab[1])
        plt.title(title[0], fontsize=title[1])
    else:
        sns_multi.set_axis_labels(xlab[0], ylab[0])
        #sns_multi.set_xlabels('Total Bill', size=10)
        #sns_multi.set_ylabels('Tip', size=10)
    
    
```

Simple Regression Plot

```python
plot1 = sns.lmplot(x='total_bill', y='tip', data=tips, size=6, aspect=1)

labels_and_title(['Total Bill', 14], ['Tip', 14], ['Tips vs Bill', 18])
#sns.set(rc={'figure.figsize': (10,8)}) ----> This does not work for lmplot, use lmplot's aspect and size 
plt.show()
```

Multi Regression Plots

```python
plot2 = sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', col='day', size=6, aspect=1,
                  markers=['o', 'x'], palette='Set1', col_wrap=2)
labels_and_title(['Total Bill'], ['Tip'], ['Tips vs Bill'], plot2)
sns.set_style('darkgrid')
```

Residual Plots - Regression residuals (distance from actual value)

```python
sns.residplot('total_bill', 'tip', data=tips, color='red')
labels_and_title(['Total Bill', 14], ['Tip', 14], ['Residuals Tip vs Bill', 18])
plt.show()
```

```python
from sklearn.metrics import jaccard_similarity_score
```

```python
a = str("100010101110111110101010")
b = str("010101101111011101000101")

a_list = list(map(int, list(a)))
b_list = list(map(int, list(b)))
```

```python
jaccard_similarity_score(a_list, b_list)
```

```python
both1 = 0
a1  = 0
b1 = 0
for i,j in zip(a_list, b_list):
    if i==1 and i==j:
        both1+=1
    if i==1:
        a1+=1
    if j==1:
        b1+=1
        
both1/(a1+b1-both1)
```
