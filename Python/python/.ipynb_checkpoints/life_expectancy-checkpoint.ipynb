```python
%matplotlib inline

import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('lifeexpectancy.csv')
df.head()
```

```python
#start simple
plt.plot(df.age, df.age + df.male, color = 'darkblue', linewidth = 2)
plt.plot(df.age, df.age + df.female, color = 'darkred', linewidth = 2)
```

```python
#add some proper limits and grid
plt.figure(figsize = [10, 10])

plt.plot(df.age, df.age+df.male, color = 'darkblue', linewidth = 2)
plt.plot(df.age, df.age + df.female, color = 'darkred', linewidth = 2)

plt.ylim(0, 105)

plt.grid(True)

plt.show()
```

```python
#add age baseline
plt.figure(figsize = [10, 10])

plt.plot(df.age, df.age+df.male, color = 'darkblue', linewidth = 2)
plt.plot(df.age, df.age + df.female, color = 'darkred', linewidth = 2)

plt.plot(df.age, color = 'black', linewidth = 2)

plt.ylim(0, 105)

plt.grid(True)

plt.show()
```

```python
#add fill for better focus
plt.figure(figsize = [10, 10])

plt.fill_between(y1=df.age, y2=df.female+df.age, x = df.age, color = 'green', alpha = 0.3)

plt.plot(df.age, df.age+df.male, color = 'darkblue', linewidth = 2)
plt.plot(df.age, df.age + df.female, color = 'darkred', linewidth = 2)

plt.plot(df.age, color = 'black', linewidth = 2)

plt.ylim(0, 105)

plt.grid(True)

plt.show()
```

```python
#reverse yaxis
plt.figure(figsize = [10, 10])

plt.fill_between(y1=df.age, y2=df.female+df.age, x = df.age, color = 'green', alpha = 0.3)

plt.plot(df.age, df.age+df.male, color = 'darkblue', linewidth = 2)
plt.plot(df.age, df.age + df.female, color = 'darkred', linewidth = 2)

plt.plot(df.age, color = 'black', linewidth = 2)

plt.ylim(0, 105)

plt.gca().invert_yaxis()

plt.grid(True)

plt.show()
```

```python
#flip axes
plt.figure(figsize = [10, 10])

plt.fill_betweenx(x1=df.age, x2=df.female+df.age, y = df.age, color = '#B0B1AC', alpha = 0.3)

plt.plot(df.age+df.male, df.age, color = '#268CA0', linewidth = 2)
plt.plot(df.age + df.female, df.age, color = '#B2534F', linewidth = 2)

plt.plot(df.age, color = '#3A3F4A', linewidth = 2)

plt.ylim(0, 100)
plt.xlim(0, 105)

plt.gca().invert_yaxis()

plt.grid(True)

plt.show()
```

```python
#add annotations
plt.figure(figsize = [10, 10])

plt.fill_betweenx(x1=df.age, x2=df.female+df.age, y = df.age, color = '#B0B1AC', alpha = 0.3)

plt.plot(df.age+df.male, df.age, color = '#268CA0', linewidth = 2)
plt.plot(df.age + df.female, df.age, color = '#B2534F', linewidth = 2)

plt.plot(df.age, color = '#3A3F4A', linewidth = 2)

plt.text(x = 6.5, y = 15, s = 'age')
plt.text(x = 37.5, y = 15, s = 'life expectancy')
plt.text(x = 70, y = 15, s = 'male')
plt.text(x = 79, y = 15, s = 'female')

plt.ylim(0, 100)
plt.xlim(0, 105)

plt.gca().invert_yaxis()

plt.grid(True, linestyle='dotted', linewidth=1)

plt.show()
```

```python
#add lines for annotations
plt.figure(figsize = [10, 10])

plt.fill_betweenx(x1=df.age, x2=df.female+df.age, y = df.age, color = '#B0B1AC', alpha = 0.3)

plt.plot(df.age+df.male, df.age, color = '#268CA0', linewidth = 2)
plt.plot(df.age + df.female, df.age, color = '#B2534F', linewidth = 2)

plt.plot(df.age, color = '#3A3F4A', linewidth = 2)

plt.text(x = 6.5, y = 15, s = 'age')
plt.text(x = 37.5, y = 15, s = 'life expectancy')
plt.text(x = 70, y = 15, s = 'male')
plt.text(x = 79, y = 15, s = 'female')

plt.hlines(xmax = 14.5, xmin = 1.5, y = 16, lw=1, color='#3A3F4A')
plt.hlines(xmax = 65.5, xmin = 18.5, y = 16, lw=1, color='#3A3F4A')

plt.ylim(0, 100)
plt.xlim(0, 105)

plt.gca().invert_yaxis()

plt.grid(True, linestyle='dotted', linewidth=1)

plt.show()
```

```python
#add title
plt.figure(figsize = [10, 10])

plt.fill_betweenx(x1=df.age, x2=df.female+df.age, y = df.age, color = '#B0B1AC', alpha = 0.3)

plt.plot(df.age+df.male, df.age, color = '#268CA0', linewidth = 2)
plt.plot(df.age + df.female, df.age, color = '#B2534F', linewidth = 2)

plt.plot(df.age, color = '#3A3F4A', linewidth = 2)

plt.text(x = 6.5, y = 15, s = 'age')
plt.text(x = 37.5, y = 15, s = 'life expectancy')
plt.text(x = 70, y = 15, s = 'male')
plt.text(x = 79, y = 15, s = 'female')

plt.hlines(xmax = 14.5, xmin = 1.5, y = 16, lw=1, color='#3A3F4A')
plt.hlines(xmax = 65.5, xmin = 18.5, y = 16, lw=1, color='#3A3F4A')

plt.title('Life expectancy by age and sex', fontdict =  {'fontsize':18, 'fontweight': 'bold'}, loc = 'left')

plt.ylim(0, 100)
plt.xlim(0, 105)

plt.gca().invert_yaxis()

plt.grid(True, linestyle='dotted', linewidth=1)

plt.show()
```

```python
#apply styling
fig, ax = plt.subplots(figsize=(10, 10))

plt.fill_betweenx(x1=df.age, x2=df.female+df.age, y = df.age, color = '#B0B1AC', alpha = 0.3)

plt.plot(df.age+df.male, df.age, color = '#268CA0', linewidth = 2)
plt.plot(df.age + df.female, df.age, color = '#B2534F', linewidth = 2)

plt.plot(df.age, color = '#3A3F4A', linewidth = 2)

plt.text(x = 6.5, y = 15, s = 'age')
plt.text(x = 37.5, y = 15, s = 'life expectancy')
plt.text(x = 70, y = 15, s = 'male', color = '#268CA0')
plt.text(x = 79, y = 15, s = 'female', color = '#B2534F')

plt.hlines(xmax = 14.5, xmin = 1.5, y = 16, lw=1, color='#3A3F4A')
plt.hlines(xmax = 65.5, xmin = 18.5, y = 16, lw=1, color='#3A3F4A')

plt.title('Life expectancy by age and sex\n', fontdict =  {'fontsize':18, 'fontweight': 'bold'}, loc = 'left')

plt.ylim(0, 100)
plt.xlim(0, 105)

ax.invert_yaxis()
ax.xaxis.tick_top()

ax.spines['bottom'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)

ax.xaxis.set_ticks_position('none')
ax.yaxis.set_ticks_position('none')

ax.yaxis.set_ticklabels([])

plt.grid(True, linestyle='dotted', linewidth=1)

plt.show()
```
